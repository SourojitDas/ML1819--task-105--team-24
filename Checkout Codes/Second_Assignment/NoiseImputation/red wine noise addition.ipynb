```python
import random
import pandas as pd
import numpy as np
import shutil
import os


def RandNoise(end, start, num): 
    res = []
    #print(start)
    #print(end)
    try: 
        int(start)
        for j in range(num): 
            res.append(round(random.uniform(start, end),2)) 
            #res.append(random.randint(start, end))
  
        return res 
    except ValueError:
        list = ['M','F','I']
        for j in range(num): 
            #res.append(round(random.uniform(start, end),3)) 
            #res.append(random.randint(start, end))
            res.append(random.choice(list))
  
        return res 
    
    
  
def getNoiseData(dataSet,percent):
    #dataSet = np.delete(dataSet,(0), axis=0)
    size = dataSet.shape
    rows = size[0]
    #print(rows)
    cols = size[1]-1
   
    num = (int)(rows*(percent/100))
    end = np.amax(dataSet,axis=0)
    start = np.amin(dataSet,axis=0)
    #print( start )
    #print( end )
    
    #print(start)
    #print(end)
    i = 0;
    random_val = []
    while i<cols:
        #print(start[i]+" "+end[i])
        random_val.append(RandNoise(start[i],end[i], num))
        #print(start[i])
        #print(end[i])
        i = i+1
    return random_val
#print(random_rows)


def Rand(start, end, num): 
    res = [] 
  
    for j in range(num): 
        res.append(random.randint(start, end)) 
  
    return res
  
def getNoisePlaces(dataSet,percent):
    size = dataSet.shape
    rows = size[0]
    num = (int)(rows*(percent/100))
    start = 0
    end = rows-1
    random_rows=Rand(start, end, num)
    return random_rows

def readData():
    #read both csv files to np array
    print("Reading Data....")

    files = ["winequality-red.csv"]
    for key in range(0, len(files)):
        orginal_dataset = pd.read_csv("../NumericData/"+files[key])
        size = orginal_dataset.shape
        rows = size[0]
        cols = size[1]-1
        
        #orginal_dataset = orginal_dataset[:int(rows/7)]
        
        #df1 = pd.DataFrame(orginal_dataset)
        #df1.to_csv("../NumericData/clean_"+files[key])
        #print("Saved New File As ../NumericData/clean_"+files[key])
        
        print("Using "+files[key])
    
        percent = [10,20,30,40,50]
        for x in percent:
            print("Adding "+str(x)+"% Noise")
            copy_dataset=orginal_dataset
            random_data = getNoiseData(copy_dataset,x)
        
            size = copy_dataset.shape
            rows = size[0]
            cols = size[1]-1
        
            print(size)
            k = 0
            while k<cols:
                random_places = getNoisePlaces(copy_dataset,x)
                #print(len(random_places))
                j = 0
                while j<len(random_places):
                    a = random_data[k][j]
                    row2change = random_places[j]
                    #print(row2change)
                    if j < 5:
                        print("Changing " +str(copy_dataset.iat[row2change,k])+" To "+str(a))
                    copy_dataset.iloc[row2change,k] = a
                    
                    #copy_dataset.itemset((j,k),a)
                    
                    #print(copy_dataset[j,k])
                    #print(random_data[k][j])
                    j = j+1
                k = k+1
            
            df = pd.DataFrame(copy_dataset)
            df.to_csv("../NumericData/"+str(x)+"_noisy_"+files[key],index=False)
            print("Saved New File As ../NumericData/"+str(x)+"_noisy_"+files[key])
    print("Noise Addition Completed")


readData()
```
