```python
import random
import pandas as pd
import numpy as np
import shutil
import os


def getGenderRandom(num): 
    res = []
    list = ['Male','Female']
    for j in range(num): 
        #res.append(round(random.uniform(start, end),3)) 
        #res.append(random.randint(start, end))
        res.append(random.choice(list))
    return res 
def getSeniorRandom(num): 
    res = []
    list = [0,1]
    for j in range(num): 
        #res.append(round(random.uniform(start, end),3)) 
        #res.append(random.randint(start, end))
        res.append(random.choice(list))
    return res 
def getYesNoRandom(num): 
    res = []
    list = ['Yes','No']
    for j in range(num): 
        #res.append(round(random.uniform(start, end),3)) 
        #res.append(random.randint(start, end))
        res.append(random.choice(list))
    return res   
def getYesNoInternetRandom(num): 
    res = []
    list = ['Yes','No','No internet service']
    for j in range(num): 
        #res.append(round(random.uniform(start, end),3)) 
        #res.append(random.randint(start, end))
        res.append(random.choice(list))
    return res  
def getMultipleRandom(num): 
    res = []
    list = ['Yes','No','No phone service']
    for j in range(num): 
        #res.append(round(random.uniform(start, end),3)) 
        #res.append(random.randint(start, end))
        res.append(random.choice(list))
    return res  
def getInternetRandom(num): 
    res = []
    list = ['DSL', 'Fiber optic', 'No']
    for j in range(num): 
        #res.append(round(random.uniform(start, end),3)) 
        #res.append(random.randint(start, end))
        res.append(random.choice(list))
    return res  
def getContractRandom(num): 
    res = []
    list = ['Month-to-month', 'One year', 'Two year']
    for j in range(num): 
        #res.append(round(random.uniform(start, end),3)) 
        #res.append(random.randint(start, end))
        res.append(random.choice(list))
    return res 
def getPaymentRandom(num): 
    res = []
    list = ['Electronic check', 'Mailed check', "Bank transfer (automatic)","Credit card (automatic)"]
    for j in range(num): 
        #res.append(round(random.uniform(start, end),3)) 
        #res.append(random.randint(start, end))
        res.append(random.choice(list))
    return res 
def getNoiseData(start, end, num): 
    res = [] 
    #print(start)
    #print(end)
    for j in range(num): 
        res.append(round(random.uniform(start, end),2)) 
    return res

def Rand(start, end, num): 
    res = [] 
    for j in range(num): 
        res.append(random.randint(start, end)) 
  
    return res
  
def getNoisePlaces(dataSet,percent):
    size = dataSet.shape
    rows = size[0]
    num = (int)(rows*(percent/100))
    start = 0
    end = rows-1
    random_rows=Rand(start, end, num)
    return random_rows

def readData():
    #read both csv files to np array
    print("Reading Data....")

    files = ["telecom.csv"]
    for key in range(0, len(files)):
        orginal_dataset = pd.read_csv("../NumericData/"+files[key])
        size = orginal_dataset.shape
        rows = size[0]
        cols = size[1]
        
        #orginal_dataset = orginal_dataset[:int(rows/7)]
        
        #df1 = pd.DataFrame(orginal_dataset)
        #df1.to_csv("../NumericData/clean_"+files[key])
        #print("Saved New File As ../NumericData/clean_"+files[key])
        
        print("Using "+files[key])
    
        percent = [10,20,30,40,50]
        for x in percent:
            print("Adding "+str(x)+"% Noise")
            copy_dataset=orginal_dataset
            #random_data = getNoiseData(copy_dataset,x)
        
            print(size)
            
            for col in copy_dataset:
                #print (col)
                if "gender" in col:
                    random_values = getGenderRandom((int)(rows*(x/100)))
                    random_places = getNoisePlaces(copy_dataset,x)
                    j = 0
                    while j<len(random_places):
                        a = random_values[j]
                        row2change = random_places[j]
                        #if j < 2:
                            #print("Changing " +str(copy_dataset.loc[j,col])+" To "+str(a))
                            
                        copy_dataset.loc[j,col] = a
                        #if j < 2:
                            #print(str(copy_dataset.loc[j,col]))
                        j = j+1
                if "SeniorCitizen" in col:
                    random_values = getSeniorRandom((int)(rows*(x/100)))
                    random_places = getNoisePlaces(copy_dataset,x)
                    j = 0
                    while j<len(random_places):
                        a = random_values[j]
                        row2change = random_places[j]
                        #if j < 2:
                            #print("Changing " +str(copy_dataset.loc[j,col])+" To "+str(a))
                            
                        copy_dataset.loc[j,col] = a
                        #if j < 2:
                            #print(str(copy_dataset.loc[j,col]))
                        j = j+1
                vallist = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']
                if any(col in s for s in vallist):
                    random_values = getYesNoRandom((int)(rows*(x/100)))
                    random_places = getNoisePlaces(copy_dataset,x)
                    j = 0
                    while j<len(random_places):
                        a = random_values[j]
                        row2change = random_places[j]
                        #if j < 2:
                            #print("Changing " +str(copy_dataset.loc[j,col])+" To "+str(a))
                            
                        copy_dataset.loc[j,col] = a
                        #if j < 2:
                            #print(str(copy_dataset.loc[j,col]))
                        j = j+1
                vallist = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport','StreamingTV','StreamingMovies']
                if any(col in s for s in vallist):
                    random_values = getYesNoInternetRandom((int)(rows*(x/100)))
                    random_places = getNoisePlaces(copy_dataset,x)
                    j = 0
                    while j<len(random_places):
                        a = random_values[j]
                        row2change = random_places[j]
                        #if j < 2:
                            #print("Changing " +str(copy_dataset.loc[j,col])+" To "+str(a))
                            
                        copy_dataset.loc[j,col] = a
                        #if j < 2:
                            #print(str(copy_dataset.loc[j,col]))
                        j = j+1
                if "MultipleLines" in col:
                    random_values = getMultipleRandom((int)(rows*(x/100)))
                    random_places = getNoisePlaces(copy_dataset,x)
                    j = 0
                    while j<len(random_places):
                        a = random_values[j]
                        row2change = random_places[j]
                        #if j < 2:
                            #print("Changing " +str(copy_dataset.loc[j,col])+" To "+str(a))
                            
                        copy_dataset.loc[j,col] = a
                        #if j < 2:
                            #print(str(copy_dataset.loc[j,col]))
                        j = j+1
                if "InternetService" in col:
                    random_values = getInternetRandom((int)(rows*(x/100)))
                    random_places = getNoisePlaces(copy_dataset,x)
                    j = 0
                    while j<len(random_places):
                        a = random_values[j]
                        row2change = random_places[j]
                        #if j < 2:
                            #print("Changing " +str(copy_dataset.loc[j,col])+" To "+str(a))
                            
                        copy_dataset.loc[j,col] = a
                        #if j < 2:
                            #print(str(copy_dataset.loc[j,col]))
                        j = j+1
                if "Contract" in col:
                    random_values = getContractRandom((int)(rows*(x/100)))
                    random_places = getNoisePlaces(copy_dataset,x)
                    j = 0
                    while j<len(random_places):
                        a = random_values[j]
                        row2change = random_places[j]
                        #if j < 2:
                            #print("Changing " +str(copy_dataset.loc[j,col])+" To "+str(a))
                            
                        copy_dataset.loc[j,col] = a
                        #if j < 2:
                            #print(str(copy_dataset.loc[j,col]))
                        j = j+1
                if "MonthlyCharges" in col:
                    #print(col)
                    mini = float(copy_dataset[col].min())
                    maxi = float(copy_dataset[col].max())
                    
                    random_values = getNoiseData(mini,maxi,(int)(rows*(x/100)))
                    #random_values = getYesNoRandom((int)(rows*(x/100)))
                    random_places = getNoisePlaces(copy_dataset,x)
                    j = 0
                    while j<len(random_places):
                        a = random_values[j]
                        row2change = random_places[j]
                        #if j < 2:
                            #print("Changing " +str(copy_dataset.loc[j,col])+" To "+str(a))
                            
                        copy_dataset.loc[j,col] = a
                        #if j < 2:
                            #print(str(copy_dataset.loc[j,col]))
                        j = j+1
                if "TotalCharges" in col:
                    #print(col)
                    mini = 18.8
                    maxi = 8684.0
                    
                    random_values = getNoiseData(mini,maxi,(int)(rows*(x/100)))
                    #random_values = getYesNoRandom((int)(rows*(x/100)))
                    random_places = getNoisePlaces(copy_dataset,x)
                    j = 0
                    while j<len(random_places):
                        if pd.isnull(copy_dataset.loc[j,col]):
                            j = j+1
                            continue
                        a = random_values[j]
                        row2change = random_places[j]
                        #if j < 2:
                            #print("Changing " +str(copy_dataset.loc[j,col])+" To "+str(a))
                            
                        copy_dataset.loc[j,col] = a
                        #if j < 2:
                            #print(str(copy_dataset.loc[j,col]))
                        j = j+1
                if "tenure" in col:
                    mini = float(copy_dataset[col].min())
                    maxi = float(copy_dataset[col].max())
                    
                    random_values = Rand(mini,maxi,(int)(rows*(x/100)))
                    #random_values = getYesNoRandom((int)(rows*(x/100)))
                    random_places = getNoisePlaces(copy_dataset,x)
                    j = 0
                    while j<len(random_places):
                        if pd.isnull(copy_dataset.loc[j,col]):
                            j = j+1
                            continue
                        a = random_values[j]
                        row2change = random_places[j]
                        #if j < 2:
                            #print("Changing " +str(copy_dataset.loc[j,col])+" To "+str(a))
                            
                        copy_dataset.loc[j,col] = a
                        #if j < 2:
                            #print(str(copy_dataset.loc[j,col]))
                        j = j+1
                        
                if "PaymentMethod" in col:
                    random_values = getPaymentRandom((int)(rows*(x/100)))
                    random_places = getNoisePlaces(copy_dataset,x)
                    j = 0
                    while j<len(random_places):
                        a = random_values[j]
                        row2change = random_places[j]
                        #if j < 2:
                            #print("Changing " +str(copy_dataset.loc[j,col])+" To "+str(a))
                            
                        copy_dataset.loc[j,col] = a
                        #if j < 2:
                            #print(str(copy_dataset.loc[j,col]))
                        j = j+1       
                            
            df = pd.DataFrame(copy_dataset)
            df.to_csv("../NumericData/"+str(x)+"_noisy_"+files[key],index=False)
            print("Saved New File As ../NumericData/"+str(x)+"_noisy_"+files[key])
    print("Noise Addition Completed")


readData()
```
